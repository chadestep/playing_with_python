# Stylistic choices for matplotlib figures

```python
%matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

## Generate test data for the basic stylesheet

```python
np.random.seed(3)
noise = (2*np.random.randn(1000))
x = np.linspace(0,100,1000)
y = np.array(10 * np.sin(x) + noise)

f, ax = plt.subplots(1)
ax.plot(x,y)
ax.set_title('Baseline matplotlib')
ax.set_ylabel('Y label')
ax.set_xlabel('X label')
plt.show()
```

### Let's see what plotting a bunch of oscillations looks like (from top to bottom)

```python
f, ax = plt.subplots(1)
for i in range(11):
    ax.plot(x,y+(i*-15))
ax.set_title('Stylesheet test')
ax.set_ylabel('Y label')
ax.set_xlabel('X label')
plt.show()
```
